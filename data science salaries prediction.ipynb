feature engineering

```python
df1=df.drop(columns=['salary_currency','salary'],axis=1)
```

```python
df1.shape
```

```python
df1.head(4)
```

```python
q1,q3=np.percentile(df['salary_in_usd'],[27,75])
```

```python
q1,q3
```

```python
iqr=q3-q1
```

```python
low_boundary=q1-(1.5*iqr)
high_boundary=q3+(1.5*iqr)    
```

```python
low_boundary,high_boundary
```

```python
df1.shape
```

```python
df2=df1[(df1['salary_in_usd']>low_boundary) & (df1['salary_in_usd']< high_boundary)]
```

```python
df2.shape
```

```python
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns # array of column names to encode

    def fit(self,X,y=None):
        return self # not relevant here

    def transform(self,X):
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
df3=MultiColumnLabelEncoder(columns = ['experience_level','employment_type','job_title','employee_residence','company_location','company_size']).fit_transform(df2)
```

```python
x=df3.drop(columns='salary_in_usd',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=80)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
```

```python
reg.fit(x_train,y_train)
```

```python
print("predicted: {0},'/n' real values: {1}".format(reg.predict(x_test),y_test))
```

```python
from sklearn.metrics import r2_score,mean_squared_error 
```
